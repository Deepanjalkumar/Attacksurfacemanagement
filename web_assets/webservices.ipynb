# Importing packages

```python
# Importing packages
import os
```

# Getting the location of working directory

```python
# Getting the location of working directory

def getting_location_of_working_directory():
    try:
        cwd=os.getcwd()
        print(f"Location of current working directory :", cwd)
    except Exception as e:
        print(e)
getting_location_of_working_directory()
```

# Making directory name Web_service_results

```python
# Making directory Web_service_results to save the results

def making_directory_Web_service_results():
    try:
        os.mkdir("Web_service_results")
    except Exception as e:
        print("Directory name Web_service_results already exists")
making_directory_Web_service_results()
```

# Performing web services check up to look for running web server

```python
def webservice_checkup():
    try:
        # Using httprobe
        os.system(f"cat Subdomain_results/all_domains.txt | httprobe > Web_service_results/httprobe.txt")
        # Using httpx   
        os.system(f"cat Subdomain_results/all_domains.txt | httpx > Web_service_results/httpx.txt")
    except Exception as e:
        print(e)
        
webservice_checkup()
```

# Sorting them in a single file

```python
def sorting_web_service_single_file():
    try:
        os.system("cat Web_service_results/httprobe.txt > Web_service_results/all_web_service.txt")
        os.system("cat Web_service_results/httpx.txt >> Web_service_results/all_web_service.txt")
        os.system("cat Web_service_results/all_web_service.txt | sort -u | tee Web_service_results/all_web_service.txt")
    except Exception as e:
        print(e)
sorting_web_service_single_file()
```
