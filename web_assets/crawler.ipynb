# Getting the location of working directory

```python
# Getting the location of working directory

def getting_location_working_directory():
    cwd=os.getcwd()
    print("Location of current working directory :", cwd)
getting_location_working_directory()
```

# Making directory name crawler

```python
# Making the directory name crawler

def directory_name_crawler():
    try:
        os.mkdir("Crawler")
    except Exception as e:
        print(e)
directory_name_crawler()
```

# Crawling over websites

```python
# Performing crawling over the input targets

def crawling():
    try:
        os.system('gospider -S Web_service_results/all_web_service.txt  -c 10 -d 1 --other-source --blacklist  ".(jpg|jpeg|gif|css|tif|tiff|png|ttf|woff|woff2|ico)" -t 10 | tee Crawler/gospider.txt')
    except Exception as e:
        print(e)
crawling()
```

# Sorting the files

```python
# Sorting the files

def sorting_files():
    try:
        os.system("cat Crawler/gospider/*/ | tee Crawler/gospider_all.txt")
        os.system("cat Crawler/gospider_all.txt | sort -u | tee Crawler/gospider_all.txt ")
    except Exception as e:
        print(e)
sorting_files()
```
