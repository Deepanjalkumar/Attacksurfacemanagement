# Getting the location of working directory

```python
# Getting the location of working directory

def getting_location_working_directory():
    try:
        cwd=os.getcwd()
        print("Location of current directory :", cwd)
    except Exception as e:
        print(e)
getting_location_working_directory()
```

# Making directory name Web_screenshot

```python
# Making directory name web_screenshot to save screenshots

def making_directory_Web_screenshot():
    try:
        os.mkdir("Web_screenshot")
    except Exception as e:
        print(e)
making_directory_Web_screenshot()
```

# Taking screenshot of the website

```python
# Taking screenshot

def taking_screenshot():
    try:
        #Using gowitness
        os.system("gowitness file --chrome-path /usr/bin/chromium -f Web_service_results/all_web_service.txt report -P Web_screenshot/ --delay 5")
    except Exception as e:
        print(e)
        
taking_screenshot()
```
