# Introduction : Subdomain enumeration

Sub-domain enumeration is the process of finding sub-domains for one or more domain(s). It is an essential part of the reconnaissance phase.

# Helper : Helpers needed for brute forcing subdomains

```python
#Some of the helpers needed for maximum covergae
resolvers="40resolvers.txt"
config="config.ini"
dnsresolvers="/root/Tools/massdns/lists/resolvers.txt"
wordlist1="/usr/share/amass/wordlists/all.txt"
wordlist2="/usr/share/amass/wordlists/subdomains.txt"
wordlist3="/usr/share/amass/wordlists/subdomains-top1mil-5000.txt"
wordlist4="/usr/share/amass/wordlists/subdomains-top1mil-20000.txt"
wordlist5="/usr/share/amass/wordlists/subdomains-top1mil-110000.txt"
wordlist6="/usr/share/amass/wordlists/subdomains.lst"
wordlist7="/usr/share/amass/wordlists/sorted_knock_dnsrecon_fierce_recon-ng.txt"
wordlist8="/usr/share/amass/wordlists/jhaddix_all.txt"
wordlist9="/usr/share/amass/wordlists/fierce_hostlist.txt"
wordlist10="/usr/share/amass/wordlists/deepmagic.com_top50kprefixes.txt"
alterationwords="/usr/share/amass/wordlists/words.txt"
chaos_api_key=""
```

```python
# Importing packages
import os
```

```python
# Making directory Subdomain_results to save the results

def making_directory_Subdomain_results():
    try:
        os.mkdir("Subdomain_results")
    except Exception as e:
        print("Directory name Subdomain_results already exists")
making_directory_Subdomain_results()
```

# Performing subdomain enumeration

# Enter the domain name 

```python
# Enter the domain name

domain=input("Enter the domain name : " )
print(f"Performin subdomain enumeration on {domain}")
```

# Using Assetfinder

```python
def subdomain_enumeration(domain):
    try:
        # Using assetfinder
        os.system(f"assetfinder -subs-only {domain} > Subdomain_results/assetfinder.txt")
    except Exception as e:
        print("Cannot run Assetfinder at this point")
        
subdomain_enumeration(domain)
```

# Using Subfinder

```python
def subdomain_enumeration(domain):
    try:
        #Using subfinder
        os.system(f"subfinder -d {domain} -nW -v -rL {resolvers} -o Subdomain_results/subfinder.txt")
    except Exception as e:
        print("Cannot run subfinder at this point")
subdomain_enumeration(domain)
```

# Using Chaos client

```python
def subdomain_enumeration(domain):
    try:
        #Using chaos client
        os.system(f"chaos -d {domain} -key {chaos_api_key} > Subdomain_results/chaos.txt")
    except Exception as e:
        print("Cannot run chaos client at this point")
subdomain_enumeration(domain)
```

# Using Metabigor

```python
def subdomain_enumeration(domain):
    try:
        #Using metabigor
        os.system(f"echo {domain} | metabigor net --domain -v -o Subdomain_results/metabigor.txt")
    except Exception as e:
        print("Cannot run metabigor at this point")
subdomain_enumeration(domain)
```

# Using Amass for passive enumeration

```python
def subdomain_enumeration(domain):
    try:
        #Using amass for passive enumeration
        os.system(f"amass enum -src -passive  -rf {resolvers} -max-dns-queries 20000 -d {domain} -o Subdomain_results/amass_enum_passive.txt")
    except Exception as e:
        print("Cannot run amass at this point")
subdomain_enumeration(domain)
```

# Using Amass for active enumeration

```python
def subdomain_enumeration(domain):
    try:
        #Using amass for active enumeration
        os.system(f"amass enum -src -active  -rf {resolvers} -max-dns-queries 20000 -d {domain} -o Subdomain_results/amass_enum_active.txt")
    except Exception as e:
        print("Cannot run amass at this point")
subdomain_enumeration(domain)
```

# Using shuffledns for subdomain enumeration via brute force 

```python
def subdomain_enumeration(domain):
    try:
        #Using shuffledns 
        os.system(f"shuffledns  -d {domain} -w {wordlist1} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain1_brute_all.txt")
        os.system(f"shuffledns  -d {domain} -w {wordlist2} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain_brute_subdomains.txt")
        os.system(f"shuffledns  -d {domain} -w {wordlist3} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain_brute_subdomains-top1mil-5000.txt")
        os.system(f"shuffledns  -d {domain} -w {wordlist4} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain_brute_subdomains-top1mil-20000.txt")
        os.system(f"shuffledns  -d {domain} -w {wordlist5} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain_brute_subdomains-top1mil-110000.txt")
        os.system(f"shuffledns  -d {domain} -w {wordlist6} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain_brute_subdomains_lst.txt")
        os.system(f"shuffledns  -d {domain} -w {wordlist7} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain_brute_sorted_knock_dnsrecon_fierce_recon-ng.txt")
        os.system(f"shuffledns  -d {domain} -w {wordlist8} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain_brute_jhaddix_all.txt")
        os.system(f"shuffledns  -d {domain} -w {wordlist9} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain_brute_fierce_hostlist.txt")
        os.system(f"shuffledns  -d {domain} -w {wordlist10} -r {dnsresolvers} -o Subdomain_results/shuffledns_domain_brute_deepmagic.com_top50kprefixes.txt")
    except Exception as e:
        print("Cannot run shuffledns at this point")
subdomain_enumeration(domain)
```

# Using goaltdns for subdomain enumeration via brute force

```python
def subdomain_enumeration(domain):
    try:
        #Using goaltdns
        os.system(f"goaltdns -h {domain} -w {alterationwords} -o Subdomain_results/goaltdns1.txt")
    except Exception as e:
        print("Cannot run goaltdns at this point")
subdomain_enumeration(domain)
```
