# Resolving domain name to ip address

```python
# Resolving domain name to ip address

def resolving_to_ip(file):
    try:
        os.system(f"massdns  -r resolvers.txt -t A  {file}  > results_resolve_massdns.txt")
        os.system(f"massdns  -r resolvers.txt -t AAAA {file}  >> results_resolve_massdns.txt")        
    except Exception as e:
        print(e)
resolving_to_ip(file)
```

# Scanning for open ports

```python
# Scanning for open ports

def open_ports():
    try:
        os.system("masscan -iL all_ipv4.txt -p 0-65535 --rate=10000  --open -oG open-ports.txt")
    except Exception as e:
        print(e)
open_ports()
```
